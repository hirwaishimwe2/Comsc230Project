```python
import pandas as pd
import numpy as np
#load the data in a dataframe
data = pd.read_csv('iris.data', names =['sepal length','sepal width', 'petal length','petal width', 'class'])
#print out the data description
description=open("iris.names")
print(description.read())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#visualization
fig, axs = plt.subplots(ncols=4, figsize=(36,6))
sns.violinplot(x="class", y="sepal length", hue="class", data=data, ax=axs[0])
sns.violinplot(x="class", y="sepal width", hue="class", data=data, ax=axs[1])
sns.violinplot(x="class", y="petal length", hue="class", data=data, ax=axs[2])
sns.violinplot(x="class", y="petal width", hue="class", data=data, ax=axs[3])

```

```python

sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(x="sepal width", y="sepal length",hue="class",  data=data, ax=ax)
```

```python
sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(x="petal width", y="petal length",hue="class",  data=data, ax=ax)
```
